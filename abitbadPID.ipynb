Check if we can reproduce the light duty from the PID parameter values and experimental values

```python
index = 0

my_phi = training_y_sub[index,:,3]
my_sp = training_sp_sub[index,:]
my_L = training_L_sub[index,:]
my_t = training_t_sub[index,:]

Kp, Ki, Kd, Kbc = pid_par

my_L_pred = np.zeros_like(my_L)

error_accum = 0
error_old = my_sp[0] - my_phi[0]

sampling_time = 0.1
control_time = 0.5

for i in range(1, len(my_L)):
    if not i % int(control_time / sampling_time) == 0:
        my_L_pred[i] = my_L_pred[i-1]
        continue

    phi_val = my_phi[i]
    sp_val = my_sp[i]
    
    error = sp_val - phi_val

    error_accum += error * sampling_time * 60

    u_compute = round(Kp * error + Ki * error_accum + Kd * (error - error_old) / (sampling_time * 60))

    u = np.clip(u_compute, 0, 800)

    bc = u - u_compute
    error_accum = error_accum + bc * Kbc

    error_old = error

    my_L_pred[i] = u

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)

ax.plot(my_t, my_L, color = 'green', linewidth = 2)
ax.plot(my_t, my_L_pred, color = 'red', linewidth = 2)
```

Check if RK4 is sufficient

```python
def RK4(ode_fun, y0, t_array, L_array):
    y_out = np.zeros((len(t_array), len(y0)))
    y_out[0,:] = y0
    for i in range(len(t_array)-1):
        L0 = L_array[i]
        dt = t_array[i+1] - t_array[i]
        k1 = ode_fun(t_array[i], y_out[i,:], L0)
        k2 = ode_fun(t_array[i] + 0.5 * dt, y_out[i,:] + 0.5 * dt * k1, L0)
        k3 = ode_fun(t_array[i] + 0.5 * dt, y_out[i,:] + 0.5 * dt * k2, L0)
        k4 = ode_fun(t_array[i] + dt, y_out[i,:] + dt * k3, L0)
        y_out[i+1,:] = y_out[i,:] + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)

    return y_out
```

```python
from scipy.integrate import solve_ivp
from khammash_repro import ode_fun, get_init_cond

init_cond = get_init_cond(L0=100)
init_cond[-1] = 0.5

t_span = np.array([0, 30]) * 60
t_eval = np.linspace(t_span[0], t_span[1], 1000)

L = 0

sol = solve_ivp(ode_fun, t_span, init_cond, method='BDF', t_eval=t_eval, args=(L,), rtol=1e-6, atol=1e-9)

max_RK4_step = 0.1

t_eval_RK4 = np.arange(t_span[0], t_span[1], max_RK4_step*60)
L_RK4 = np.ones(len(t_eval_RK4)) * L

sol_RK4 = RK4(ode_fun, init_cond, t_eval_RK4, L_RK4)
```

```python
fig = plt.figure(figsize=(8, 7))
ax = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax.plot(sol.t/60, sol.y[0,:], color = 'black', linewidth = 2, label = 'Tt')
ax.plot(t_eval_RK4/60, sol_RK4[:,0], color = 'red', linewidth = 2, label = 'Tt RK4')

ax2.plot(sol.t/60, sol.y[1,:], color = 'black', linewidth = 2, label = 'Td')
ax2.plot(t_eval_RK4/60, sol_RK4[:,1], color = 'red', linewidth = 2, label = 'Td RK4')

ax3.plot(sol.t/60, sol.y[2,:], color = 'black', linewidth = 2, label = 'C')
ax3.plot(t_eval_RK4/60, sol_RK4[:,2], color = 'red', linewidth = 2, label = 'C RK4')

ax4.plot(sol.t/60, sol.y[3,:], color = 'black', linewidth = 2, label = 'phi_p')
ax4.plot(t_eval_RK4/60, sol_RK4[:,3], color = 'red', linewidth = 2, label = 'phi_p RK4')
```
